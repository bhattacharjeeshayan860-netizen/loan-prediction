# Encoding

````python
from sklearn.preprocessing import LabelEncoder
````

````python
le=LabelEncoder()
````

````python
bin_colmns=["Gender","Married","Education","Self_Employed","Loan_Status"]
````

````python
for col in bin_colmns:
    df[col]=le.fit_transform(df[col])
````

````python
df=pd.get_dummies(df,columns=["Property_Area"])
````

````python
df.head()
````

````python
cols=[col for col in df.columns if "Property_Area" in col]
df[cols]=df[cols].astype(int)
````
