# Encoding

````python
from sklearn.preprocessing import LabelEncoder
````

````python
le=LabelEncoder()
````

````python
bin_colmns=["Gender","Married","Education","Self_Employed","Loan_Status"]
````

````python
for col in bin_colmns:
    df[col]=le.fit_transform(df[col])
````

````python
df=pd.get_dummies(df,columns=["Property_Area"])
````

````python
df.head()
````

````python
cols=[col for col in df.columns if "Property_Area" in col]
df[cols]=df[cols].astype(int)
````

````python
df.head()
````

````python
df=df.drop(columns="Loan_ID")
````

````python
x=df.drop(columns="Loan_Status")
y=df["Loan_Status"]
````

## Train-Test Split
Data split into 80% train, 20% test using stratify to maintain class balance.

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,
                                               y,
                                               test_size=0.2,
                                              random_state=42,
                                              stratify=y)
````

````python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
````

## Logistic Regression
Baseline model for binary classification.
- Balanced & interpretable
- Uses class_weight to handle imbalance

````python
#using logistic regression
model=LogisticRegression(max_iter=1000,
                        solver="liblinear",
                        class_weight="balanced")
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
````

## Random Forest
Captures non-linear patterns.
- More powerful than Logistic Regression
- Slightly higher accuracy expected

````python
#using random forest classifier
model_random=RandomForestClassifier(n_estimators=200,
                                   max_depth=10,
                                   min_samples_split=5,
                                   min_samples_leaf=2,
                                   random_state=42,
                                   class_weight="balanced")
model_random.fit(x_train,y_train)
y_ran_pred=model_random.predict(x_test)
````

## Accuracy Comparison
- Logistic: ~80%
- Random Forest: ~83%
- Random Forest performs slightly better.

````python
from sklearn.metrics import accuracy_score
print("logistic: ",accuracy_score(y_test,y_pred))
print("Random Forest: ",accuracy_score(y_test,y_ran_pred))
````

## Confusion Matrix Insight
- Logistic → fewer false approvals (safer)
- Random Forest → more approvals but higher risk

````python
from sklearn.metrics import confusion_matrix
print("logistic: \n",confusion_matrix(y_test,y_pred))
print("Random Forest: \n",confusion_matrix(y_test,y_ran_pred))
````

## Precision & Recall
- Logistic → balanced
- Random Forest → higher recall (approves more), slightly lower precision

````python
from sklearn.metrics import classification_report
print("Logistic: \n",classification_report(y_test,y_pred))
print("Random Forest: \n",classification_report(y_test,y_ran_pred))
````

## Feature Importance
Credit_History is the most important feature → aligns with real-world logic.

````python
feature_importance=pd.Series(model_random.feature_importances_,index=x.columns)
print(feature_importance.sort_values(ascending=False))
````

## Overfitting Check
Train accuracy > Test accuracy → slight overfitting present.

````python
print("Train Accuract: ",model_random.score(x_train,y_train))
print("Train Accuract: ",model_random.score(x_test,y_test))
````

# Final Conclusion- Logistic → safer model- Random Forest → higher accuracy but more risk
Model choice depends on business goal.
